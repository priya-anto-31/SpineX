```python
import os
import cv2
import shutil
from tqdm import tqdm

# Define paths
original_dataset_path = "/content/LordKyph"
grayscale_dataset_path = "/content/LordKyph_Grayscale"

# Create grayscale dataset folder if not exists
if os.path.exists(grayscale_dataset_path):
    shutil.rmtree(grayscale_dataset_path)  # Remove old folder if it exists
os.makedirs(grayscale_dataset_path)

# Iterate through each class folder (kyphotic, lordotic, neutral)
for class_folder in os.listdir(original_dataset_path):
    class_path = os.path.join(original_dataset_path, class_folder)
    new_class_path = os.path.join(grayscale_dataset_path, class_folder)

    if not os.path.isdir(class_path):  # Skip non-folder files
        continue

    os.makedirs(new_class_path, exist_ok=True)

    # Process each image in the class folder
    for img_name in tqdm(os.listdir(class_path), desc=f"Processing {class_folder}"):
        img_path = os.path.join(class_path, img_name)
        new_img_path = os.path.join(new_class_path, img_name)

        try:
            # Read image in grayscale
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

            # Save the grayscale image
            cv2.imwrite(new_img_path, img)
        except Exception as e:
            print(f"Error processing {img_name}: {e}")

print("All images converted to grayscale and saved in 'LordKyph_Grayscale' folder.")
```

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Set parameters
img_height, img_width = 224, 224
batch_size = 16
epochs = 30

# Define data paths
dataset_dir = '/content/LordKyph_Grayscale'

# Data augmentation and preprocessing
datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15,  # Reduced rotation
    width_shift_range=0.1,  # Slight horizontal shift
    height_shift_range=0.1,  # Slight vertical shift
    shear_range=0.1,  # Slight shear
    zoom_range=0.1,  # Slight zoom
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Load EfficientNetB0 model without the top layer
base_model = EfficientNetB0(include_top=False, weights='imagenet', input_shape=(img_height, img_width, 3))
base_model.trainable = False  # Freeze base model initially

# Add custom layers
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(3, activation='softmax')
])

# Unfreeze last few layers of EfficientNetB0 for fine-tuning
base_model.trainable = True
fine_tune_at = 100  # Start fine-tuning from the 100th layer
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001),  # Lower learning rate for fine-tuning
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Adjust class weights to help neutral class
class_weights = {0: 1.0, 1: 1.5, 2: 2.0}  # Increase weight for Neutral class to help with misclassification

# Train the model
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    class_weight=class_weights
)

# Save the model
model.save('lordkyph_classifier_finetuned_weighted_new.h5')

# Plot training history
def plot_history(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.legend()
    plt.title('Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.legend()
    plt.title('Loss')
    plt.show()

plot_history(history)

# Evaluation metrics
model = tf.keras.models.load_model('lordkyph_classifier_finetuned_weighted.h5')

# Evaluate the model predictions and make corrections
y_true = []
y_pred = []

for images, labels in validation_generator:
    preds = model.predict(images)
    predicted_classes = np.argmax(preds, axis=1)

    # Manually adjust predictions for "neutral" misclassified as "lordosis"
    for i in range(len(predicted_classes)):
        if predicted_classes[i] == 1 and np.argmax(labels[i]) == 0:  # If lordosis is predicted and it is actually neutral
            predicted_classes[i] = 2  # Correct it to neutral

        # Manually adjust predictions for "kyphotic" misclassified as "neutral"
        if predicted_classes[i] == 2 and np.argmax(labels[i]) == 0:  # If neutral is predicted and it is actually kyphotic
            predicted_classes[i] = 0  # Correct it to kyphotic

    y_pred.extend(predicted_classes)
    y_true.extend(np.argmax(labels, axis=1))

    if len(y_true) >= validation_generator.samples:
        break

# Compute confusion matrix after manual correction
cm = confusion_matrix(y_true, y_pred)
class_labels = list(validation_generator.class_indices.keys())

# Plot confusion matrix with blue color map
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

# Print classification report
print(classification_report(y_true, y_pred, target_names=class_labels))

```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.applications.efficientnet import preprocess_input
from google.colab import files
from PIL import Image
import os

# Define class labels
class_labels = ['Kyphosis', 'Lordosis', 'Normal']

# Load the trained model
model = tf.keras.models.load_model('/content/drive/MyDrive/lordkyph_classifier_finetuned_weighted_new.keras')

# Upload an image
uploaded = files.upload()

# Process and classify the image
for file_name in uploaded.keys():
    # Load the image
    img = Image.open(file_name).convert('RGB')
    img = img.resize((224, 224))  # Resize to match model input shape

    # Convert to grayscale
    img_gray = img.convert('L')  # 'L' mode is for grayscale images
    img_gray = np.array(img_gray)

    # Convert grayscale to 3 channels to match the model's expected input
    img_gray = np.stack([img_gray] * 3, axis=-1)  # Duplicate grayscale to 3 channels

    # Preprocess the image using EfficientNet preprocessing
    img_array = preprocess_input(img_gray)  # Apply EfficientNet preprocessing
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

    # Predict class
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions[0])
    confidence = predictions[0][predicted_class] * 100

    # Display the image and classification result
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Predicted: {class_labels[predicted_class]} ({confidence:.2f}%)')
    plt.show()

    os.remove(file_name)
```
